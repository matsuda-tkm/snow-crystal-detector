## まずは動かしてみよう

1. `coco_dataset/`配下に`train.txt`, `val.txt`, `test.txt`を作成
2. 各`.txt`ファイルには`coco_dataset/images/`配下の画像ファイルのパスを記載（例：`coco_dataset/images/XXXXXXX.png`）
3. `coco.yaml`を指定して学習を実行（`.train`メソッド）
4. テストデータで評価する（`.val`メソッド）

```python
from ultralytics import YOLO

# モデルを読み込む
model = YOLO("yolo11n.yaml").load("yolo11n.pt")

# モデルを訓練する
results = model.train(
    data="coco.yaml",
    epochs=10,
    imgsz=360,
    plots=True,
    name="example",
)

# モデルを検証する
model.val(split="val")
model.val(split="test")
```
